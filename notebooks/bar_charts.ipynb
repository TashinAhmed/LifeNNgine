```python
import os
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Learning Life with different activation functions.
```

```python
filenames = [
        os.path.join("results", "poly_l1_1", "exp.csv"),
        os.path.join("results", "prelu_l1_1", "exp.csv"),
        os.path.join("results", "square_l1_1", "exp.csv"),
        os.path.join("results", "silu_l1_1", "exp.csv"),
        os.path.join("results", "rootsquare_l1_1", "exp.csv"),
        os.path.join("results", "leakyrelu_l1_1", "exp.csv"),
        os.path.join("results", "celu_l1_1", "exp.csv"),
        os.path.join("results", "sigmoid_l1_1", "exp.csv"),
        os.path.join("results", "tanh_l1_1", "exp.csv"),
        os.path.join("results", "relu_l1_1", "exp.csv")]

fig, ax = plt.subplots(1,1, figsize=(8,4))
my_cmap = plt.get_cmap("plasma")
labels = []
label_locs = []

for idx, fn in enumerate(filenames):

    df = pd.read_csv(os.path.join("..",fn))

    accuracy = (df["final_grid_accuracy"] == 1.0).mean()
    label = os.path.split(os.path.split(fn)[0])[1]

    p = ax.bar(1.5*idx, accuracy, color=my_cmap(16*idx), label=label)
    ax.bar_label(p, label_type="edge") #f"{accuracy:.3f}")
    label_locs.append(1.5*idx-0.75)
    
    labels.append(label)

    
    
    #plt.legend()
ax.axis([-.5, idx*1.5+1., 0, 1.1])
ax.set_xticks(label_locs)
ax.set_xticklabels(labels, rotation=45)
ax.set_title("Success rates by activation function\nL(1,1) networks, 16 runs")
plt.savefig("activations.png")
plt.show()
```

```python
# Sensitivity to initial conditions density. 
```

```python
filenames = os.listdir("../results")

for fn in filenames:

    if "density" in fn:
        df = pd.read_csv(os.path.join("..","results",fn, "exp.csv"), index_col=0)
        my_title = fn #os.path.split(os.path.split(fn)[0])[1]
        
        fig, ax = plt.subplots(1,1, figsize=(8,4))
        densities = df["density_0"].unique()
        labels = []
        label_locs = []
        bar_width = len(densities) // 3
        for idx, d0 in enumerate(densities):
            
            accuracy = (df[df["density_0"]==d0]["final_grid_accuracy"] == 1.0).mean()
            label = f"{d0:.3f}"

            p = ax.bar(bar_width*1.5*idx, accuracy, color=my_cmap(16*idx), label=label, width=bar_width)
            #ax.bar_label(p, label_type="edge", rotation=90)
            label_locs.append((1.5*bar_width)*idx)
            
            labels.append(label)
        
        ax.plot([-bar_width, (len(densities))*1.5*bar_width], [1.,1.], "--", color="k", alpha=0.125, label="1.0")
        
        ax.axis([-bar_width, (len(densities))*1.5*bar_width, 0, 1.1])
        ax.set_xticks(label_locs)
        ax.set_xticklabels(labels, rotation=60)
        ax.set_title(f"Success rates by density\n {my_title}")
        plt.savefig(f"{my_title}.png")
        plt.show()
            
```
