```python
import matplotlib.pyplot as plt

import matplotlib.gridspec as gridspec


def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        ax.tick_params(labelbottom=False, labelleft=False)


# gridspec inside gridspec
fontsize = 16

fig = plt.figure(figsize=(13,13))

gs0 = gridspec.GridSpec(2, 2, figure=fig)


gs00 = gridspec.GridSpecFromSubplotSpec(2, 2, subplot_spec=gs0[0,0],\
      width_ratios=[0.8,0.2],\
      height_ratios=[0.8, 0.2])

gs01 = gridspec.GridSpecFromSubplotSpec(2, 2, subplot_spec=gs0[0,1],\
      width_ratios=[0.8,0.2],\
      height_ratios=[0.8, 0.2])


gs10 = gridspec.GridSpecFromSubplotSpec(2, 2, subplot_spec=gs0[1,0],\
      width_ratios=[0.8,0.2],\
      height_ratios=[0.8, 0.2])


gs11 = gridspec.GridSpecFromSubplotSpec(2, 2, subplot_spec=gs0[1,1],\
      width_ratios=[0.8,0.2],\
      height_ratios=[0.8, 0.2])


ax_main00 = fig.add_subplot(gs00[0, 0])
ax_projv00 = fig.add_subplot(gs00[0, 1])
ax_projh00 = fig.add_subplot(gs00[1, 0])

ax_main01 = fig.add_subplot(gs01[0, 0])
ax_projv01 = fig.add_subplot(gs01[0, 1])
ax_projh01 = fig.add_subplot(gs01[1, 0])

ax_main10 = fig.add_subplot(gs10[0, 0])
ax_projv10 = fig.add_subplot(gs10[0, 1])
ax_projh10 = fig.add_subplot(gs10[1, 0])

ax_main11 = fig.add_subplot(gs11[0, 0])
ax_projv11 = fig.add_subplot(gs11[0, 1])
ax_projh11 = fig.add_subplot(gs11[1, 0])

my_path00 = os.path.join("..", "results", "poly_weight_knockout_neur_l1_1", "exp.csv")
my_path01 = os.path.join("..", "results", "prelu_weight_knockout_neur_l1_1", "exp.csv")
my_path10 = os.path.join("..", "results", "relu_no_knockout_l1_1", "exp.csv")
my_path11 = os.path.join("..", "results", "sigmoid_no_knockout_l1_1", "exp.csv")

df11 = pd.read_csv(my_path11)
df10 = pd.read_csv(my_path10)
df01 = pd.read_csv(my_path01)
df00 = pd.read_csv(my_path00)

plot_traj_proj_on_axes(ax_main00, ax_projv00, ax_projh00, df00, main_title="PolyKAN L(1,1) (static neurons)",\
        fontsize=fontsize, marker_size=8, small_marker_size=2,\
        clear_labels=[[0,0,1],[1,1,0],[0,0,1]])
plot_traj_proj_on_axes(ax_main01, ax_projv01, ax_projh01, df01, main_title="PReLU L(1,1) (static neurons)",\
        fontsize=fontsize, marker_size=8, small_marker_size=2, \
        clear_labels=[[0,1,1],[1,1,0],[1,1,1]])
plot_traj_proj_on_axes(ax_main10, ax_projv10, ax_projh10, df10, main_title="ReLU L(1,16)",\
        fontsize=fontsize, marker_size=8, small_marker_size=2,\
        clear_labels=[[0,0,1],[0,0,0],[0,0,0]])
plot_traj_proj_on_axes(ax_main11, ax_projv11, ax_projh11, df11, main_title="Sigmoid L(1,1)",\
        fontsize=fontsize, marker_size=8, small_marker_size=2,\
        clear_labels=[[0,0,1],[0,0,0],[0,0,0]])

fig.suptitle(f"PCA Parameter Space Trajectories", fontsize=fontsize+4)

plt.tight_layout()
plt.savefig("pca_param_space.png")
plt.show()
```
