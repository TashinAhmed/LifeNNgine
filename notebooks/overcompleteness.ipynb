```python
filenames = [os.path.join("results", "poly_l1_m", "exp.csv"),\
             os.path.join("results", "prelu_l1_m", "exp.csv"),\
             os.path.join("results", "silu_l1_m", "exp.csv"),\
             os.path.join("results", "relu_l1_m", "exp.csv")
            ]

fig, ax = plt.subplots(1,1, figsize=(7,5))
my_cmap = plt.get_cmap("plasma")
labels = []
label_locs = []
bar_width = 2.
color_coeff = 32

for idx, fn in enumerate(filenames):

    df = pd.read_csv(os.path.join("..",fn))

    for idx_2, width in enumerate(df["model_width"].unique()):
        
        accuracy = (df[df["model_width"]==width]["final_grid_accuracy"] == 1.0).mean()
        label = os.path.split(os.path.split(fn)[0])[1]
    
        p = ax.bar(idx_2*bar_width+4*bar_width*idx, accuracy, width=bar_width, color=my_cmap(color_coeff*(idx+idx_2)), label=label)
        ax.bar_label(p, labels=[f"{accuracy:.4f}"],label_type="edge", rotation=75) #f"{accuracy:.3f}")

        if accuracy < 0.5:
            ax.bar_label(p,  labels=[f"\nm = {width}"], fontsize=12, label_type="center", color="black", rotation=75) 
        else:
            ax.bar_label(p,  labels=[f"m = {width}"], fontsize=11, label_type="center", color="white", rotation=60) 
        
        label_locs.append((4*bar_width*idx)+bar_width)
        
        labels.append(label)

    
    
    #plt.legend()
ax.axis([-bar_width, bar_width+idx_2*bar_width+4*bar_width*idx, 0, 1.4])
ax.set_xticks(label_locs)
ax.set_xticklabels(labels, rotation=45)
ax.set_title("Success rates by overcompleteness $m = (2,4,8)$")
plt.tight_layout()
plt.savefig("overcompleteness.png")
plt.show()
```
