```python
import os
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
```

```python

my_cmap = plt.get_cmap("plasma")
filenames = os.listdir("../results")

filenames = [\
        os.path.join("..","results", "density_polykan_l1_1", "exp.csv"),\
        os.path.join("..","results", "density_polykan_l1_1_zoom", "exp.csv"),\
        os.path.join("..","results", "density_polykan_l1_2", "exp.csv"),\
        os.path.join("..","results", "density_prelu_l1_1", "exp.csv"),\
        os.path.join("..","results", "density_prelu_l1_1_zoom", "exp.csv"),\
        os.path.join("..","results", "density_prelu_l1_2", "exp.csv"),\
        os.path.join("..","results", "density_silu_l1_1", "exp.csv"),\
        os.path.join("..","results", "density_silu_l1_1_zoom", "exp.csv"),\
        os.path.join("..","results", "density_silu_l1_2", "exp.csv"),\
        os.path.join("..","results", "density_relu_l1_1", "exp.csv"),\
        os.path.join("..","results", "density_relu_l1_1_zoom", "exp.csv"),\
        os.path.join("..","results","density_relu_l1_2", "exp.csv")]
#    os.path.join("..","results","density_relu_128runs_l1_1", "exp.csv")]

axes_indices = [[0,0], [0,1], [0,2],\
                [1,0], [1,1], [1,2],\
                [2,0], [2,1], [2,2],\
                [3,0], [3,1], [3,2]]

titles = ["PolyKAN $L(1,1)$", "PolyKAN $L(1,1)$\nzoom", "PolyKAN $L(1,2)$",\
          "PReLU $L(1,1)$", "PReLU $L(1,1)$\nzoom", "PReLU $L(1,2)$",\
          "SiLU $L(1,1)$", "SiLU $L(1,1)$\nzoom", "SiLU $L(1,2)$",\
          "ReLU $L(1,1)$", "ReLU $L(1,1)$\nzoom", "ReLU $L(1,2)$"
         ]

fig, ax = plt.subplots(4, 3, figsize=(14,16))
my_cmap = plt.get_cmap("plasma")

for ids, fn, title in zip(axes_indices, filenames, titles):

    if ids[1] == 0:
        ax[ids[0],ids[1]].set_ylabel("Success rate")
    if ids[0] == 3:
        ax[ids[0],ids[1]].set_xlabel("Initial 'on' density")
        
    if os.path.exists(fn):
        df = pd.read_csv(fn)
        
        my_title = fn #os.path.split(os.path.split(fn)[0])[1]
        
        densities = df["density_0"].unique()
        labels = []
        label_locs = []
        bar_width = len(densities) // 3
        
        for idx, d0 in enumerate(densities):
            
            accuracy = (df[df["density_0"]==d0]["final_grid_accuracy"] == 1.0).mean()
            label = f"{d0:.3f}"

            p = ax[ids[0],ids[1]].bar(bar_width*1.5*idx, accuracy, color=my_cmap(16*idx), label=label, width=bar_width)
            #ax.bar_label(p, label_type="edge", rotation=90)
            label_locs.append((1.5*bar_width)*idx)
            
            labels.append(label)
            ax[ids[0],ids[1]].plot([-bar_width, (len(densities))*1.5*bar_width], [1.,1.], "--", color="k", alpha=0.125, label="1.0")
            
            if ids[0] == 3:
                ax[ids[0],ids[1]].axis([-bar_width, (len(densities))*1.5*bar_width, 0, np.max(accuracy)+.3])
            else:
                ax[ids[0],ids[1]].axis([-bar_width, (len(densities))*1.5*bar_width, 0, 1.1])
            ax[ids[0],ids[1]].set_xticks(label_locs[::2])
            ax[ids[0],ids[1]].set_xticklabels(labels[::2], rotation=60)
            ax[ids[0],ids[1]].set_title(f"{title}\n{len(df[df['density_0']==d0])} runs")
            

fig.suptitle("Training success dependence on initial density\n\n", fontsize=24)
plt.tight_layout()
plt.savefig("densities_fig.png")
plt.show()
        
```

```python

my_cmap = plt.get_cmap("plasma")
filenames = os.listdir("../results")

for fn in filenames:

    if "density" in fn:
        df = pd.read_csv(os.path.join("..","results",fn, "exp.csv"), index_col=0)
        my_title = fn #os.path.split(os.path.split(fn)[0])[1]
        
        fig, ax = plt.subplots(1,1, figsize=(8,4))
        densities = df["density_0"].unique()
        labels = []
        label_locs = []
        bar_width = len(densities) // 3
        for idx, d0 in enumerate(densities):
            
            accuracy = (df[df["density_0"]==d0]["final_grid_accuracy"] == 1.0).mean()
            label = f"{d0:.3f}"

            p = ax.bar(bar_width*1.5*idx, accuracy, color=my_cmap(16*idx), label=label, width=bar_width)
            #ax.bar_label(p, label_type="edge", rotation=90)
            label_locs.append((1.5*bar_width)*idx)
            
            labels.append(label)
        
        ax.plot([-bar_width, (len(densities))*1.5*bar_width], [1.,1.], "--", color="k", alpha=0.125, label="1.0")
        
        ax.axis([-bar_width, (len(densities))*1.5*bar_width, 0, 1.1])
        ax.set_xticks(label_locs)
        ax.set_xticklabels(labels, rotation=60)
        ax.set_title(f"Success rates by density\n {my_title}")
        plt.tight_layout()
        plt.savefig(f"{my_title}.png")
        plt.show()
            
```
