```python
import os
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Learning Life with different activation functions.
```

```python
filenames = [
        os.path.join("results", "poly_l1_1", "exp.csv"),
        os.path.join("results", "prelu_l1_1", "exp.csv"),
        os.path.join("results", "square_l1_1", "exp.csv"),
        os.path.join("results", "silu_l1_1", "exp.csv"),
        os.path.join("results", "rootsquare_l1_1", "exp.csv"),
        os.path.join("results", "leakyrelu_l1_1", "exp.csv"),
        os.path.join("results", "celu_l1_1", "exp.csv"),
        os.path.join("results", "sigmoid_l1_1", "exp.csv"),
        os.path.join("results", "tanh_l1_1", "exp.csv"),
        os.path.join("results", "relu_l1_1", "exp.csv")]

fig, ax = plt.subplots(1,1, figsize=(8,5))
my_cmap = plt.get_cmap("plasma")
labels = []
label_locs = []

for idx, fn in enumerate(filenames):

    df = pd.read_csv(os.path.join("..",fn))

    accuracy = (df["final_grid_accuracy"] == 1.0).mean()
    label = os.path.split(os.path.split(fn)[0])[1]

    p = ax.bar(1.5*idx, accuracy, color=my_cmap(16*idx), label=label)
    ax.bar_label(p, labels=[f"{accuracy:.3f}"], label_type="edge", rotation=60) #f"{accuracy:.3f}")
    label_locs.append(1.5*idx-0.75)
    
    labels.append(label)

    
    number_runs = len(df)
    #plt.legend()
ax.axis([-.5, idx*1.5+1., 0, 1.3])
ax.set_xticks(label_locs)
ax.set_xticklabels(labels, rotation=45)
ax.set_title(f"Success rates by activation function\nL(1,1) networks, {number_runs} runs")
plt.tight_layout()
plt.savefig("minimal_models_by_activation.png")
plt.show()
```
