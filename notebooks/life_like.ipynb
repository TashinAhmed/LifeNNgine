```python
filenames = [
        os.path.join("results", "poly_l1_m_dotlife", "exp.csv"),
        os.path.join("results", "poly_l1_m_morley", "exp.csv"),
        os.path.join("results", "poly_l1_m_fredkin", "exp.csv"),
        os.path.join("results", "poly_l1_m_asal_d", "exp.csv"),
        os.path.join("results", "prelu_l1_m_dotlife", "exp.csv"),
        os.path.join("results", "prelu_l1_m_morley", "exp.csv"),
        os.path.join("results", "prelu_l1_m_fredkin", "exp.csv"),
        os.path.join("results", "prelu_l1_m_asal_d", "exp.csv"),
        os.path.join("results", "silu_l1_m_dotlife", "exp.csv"),
        os.path.join("results", "silu_l1_m_morley", "exp.csv"),
        os.path.join("results", "silu_l1_m_fredkin", "exp.csv"),
        os.path.join("results", "silu_l1_m_asal_d", "exp.csv"),
        os.path.join("results", "relu_l1_m_dotlife", "exp.csv"),
        os.path.join("results", "relu_l1_m_morley", "exp.csv"),
        os.path.join("results", "relu_l1_m_fredkin", "exp.csv"),
        os.path.join("results", "relu_l1_m_asal_d", "exp.csv")
        ]

my_cmap = plt.get_cmap("plasma")
labels = []
label_locs = []
bar_width = 5
bar_width_b = 1

model_widths = [1,2,4,8]
model_labels = ["PolyKAN",\
                "PolyKAN",\
                "PolyKAN",\
                "PolyKAN",\
                "PReLU",\
                "PReLU",\
                "PReLU",\
                "PReLU",\
                "SiLU",\
                "SiLU",\
                "SiLU",\
                "SiLU",\
                "ReLU",\
                "ReLU",\
                "ReLU",\
                "ReLU"]


rule_labels = ["DotLife",\
                "Morley",\
                "Fredkin",\
                "ASAL d"
              ]


fig, ax = plt.subplots(4,1, figsize=(12,16))

for ax_idx in range(4):
    for idx, fn in enumerate(filenames[4*ax_idx:4*(1+ax_idx)]):
        
        df = pd.read_csv(os.path.join("..",fn))
        
        for idx_b, width in enumerate(model_widths):
            
            sub_df = df[df["model_width"] == width]
            accuracy = (sub_df["final_grid_accuracy"] == 1.0).mean()
            my_total = len(sub_df["final_grid_accuracy"])
            my_sum = ((sub_df["final_grid_accuracy"]==1.0).sum())
                  
            p = ax[ax_idx].bar(bar_width*idx+bar_width_b*idx_b, accuracy, width=bar_width_b, color=my_cmap(48*idx+idx_b*12), label=model_labels[idx])
            
            ax[ax_idx].bar_label(p, labels=[f"  {my_sum}/{my_total}"], label_type="edge", rotation=90) #f"{accuracy:.3f}")
            label_locs.append(bar_width*idx+bar_width_b*idx_b)

            if idx_b>10:
                labels.append(f"L(1,{width})")
            else:
                labels.append(f"{rule_labels[idx%4]}\n L(1,{width})")
        
    ax[ax_idx].axis([-bar_width_b, (idx+1)*bar_width, 0, 1.3])
    ax[ax_idx].set_xticks(label_locs)
    ax[ax_idx].set_xticklabels(labels, rotation=90)
    ax[ax_idx].set_title(f"Learning Life-like CA\n{model_labels[ax_idx*4]}")
    
plt.tight_layout()
plt.savefig("life_like.png")
plt.show()
```
